```python
import random

# Input ukuran peta dari pengguna
baris = int(input("Masukkan jumlah baris peta: "))
kolom = int(input("Masukkan jumlah kolom peta: "))

# Membuat peta tersembunyi dan peta tampilan untuk pemain
peta_asli = [["-" for _ in range(kolom)] for _ in range(baris)]
peta_tampil = [["?" for _ in range(kolom)] for _ in range(baris)]  # Pemain hanya melihat tanda "?"

# Mengacak lokasi harta karun
harta_x, harta_y = random.randint(0, baris - 1), random.randint(0, kolom - 1)
peta_asli[harta_x][harta_y] = "X"

# Menampilkan peta awal
print("\nPeta Treasure Hunt (Tebak Lokasi Harta Karun!):")
for row in peta_tampil:
    print(" ".join(row))

# Pemain menebak lokasi harta karun
while True:
    tebakan_x = input("\nMasukkan tebakan baris (atau 'q' untuk keluar): ")

    if tebakan_x.lower() == "q":
        print("\nPermainan dihentikan. Sampai jumpa!")
        break

    tebakan_y = input("Masukkan tebakan kolom (atau 'q' untuk keluar): ")

    if tebakan_y.lower() == "q":
        print("\nPermainan dihentikan. Sampai jumpa!")
        break

    # Memvalidasi input untuk menerima angka dalam rentang yang benar
    if not tebakan_x.isdigit() or not tebakan_y.isdigit():
        print("\nMasukkan angka yang valid!")
        continue

    tebakan_x, tebakan_y = int(tebakan_x) - 1, int(tebakan_y) - 1

    if tebakan_x < 0 or tebakan_x >= baris or tebakan_y < 0 or tebakan_y >= kolom:
        print("\nTebakan di luar batas! Coba lagi.")
        continue

    if tebakan_x == harta_x and tebakan_y == harta_y:
        print("\nSelamat! Anda menemukan harta karun!")
        peta_tampil[harta_x][harta_y] = "X"
        break
    else:
        print("\nBelum ketemu, coba lagi!")
        peta_tampil[tebakan_x][tebakan_y] = "O"  # Menandai lokasi tebakan yang salah

    # Menampilkan peta setelah tebakan
    for row in peta_tampil:
        print(" ".join(row))

# Menampilkan peta akhir dengan lokasi harta karun
print("\nPeta Akhir:")
for row in peta_tampil:
    print(" ".join(row))
```
