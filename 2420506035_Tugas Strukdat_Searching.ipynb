```python
import random
import math
import pandas as pd

def generate_sorted_array():
    return sorted(random.sample(range(1, 1001), 100))

def linear_search(arr, target):
    counter = 0
    for i in range(len(arr)):
        counter += 1
        if arr[i] == target:
            return i, counter
    return -1, counter

def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    counter = 0
    while left <= right:
        mid = (left + right) // 2
        counter += 1
        if arr[mid] == target:
            return mid, counter
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1, counter

def jump_search(arr, target):
    n = len(arr)
    step = int(math.sqrt(n))
    prev = 0
    counter = 0

    while prev < n and arr[min(step, n) - 1] < target:
        counter += 1
        prev = step
        step += int(math.sqrt(n))
        if prev >= n:
            return -1, counter

    for i in range(prev, min(step, n)):
        counter += 1
        if arr[i] == target:
            return i, counter
    return -1, counter

# Buat daftar angka terurut
arr = generate_sorted_array()

# Tentukan target
target = arr[len(arr) // 2]

# Jalankan pencarian
hasil_lin, hitung_lin = linear_search(arr, target)
hasil_bin, hitung_bin = binary_search(arr, target)
hasil_lompat, hitung_lompat = jump_search(arr, target)

# Buat DataFrame untuk menampilkan hasil
hasil_df = pd.DataFrame({
    "Algoritma Pencarian": ["Linear Search", "Binary Search", "Jump Search"],
    "Jumlah Perbandingan": [hitung_lin, hitung_bin, hitung_lompat]
})

# Cetak tabel hasil
print(hasil_df)
```
