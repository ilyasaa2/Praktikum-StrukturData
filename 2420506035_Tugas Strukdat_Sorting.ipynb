1. Optimasi Pencarian Produk di Marketplace (Rombel 02)

```python
def quick_sort(arr, key):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x[key] < pivot[key]]
    middle = [x for x in arr if x[key] == pivot[key]]
    right = [x for x in arr if x[key] > pivot[key]]
    return quick_sort(left, key) + middle + quick_sort(right, key)

# Contoh data produk
produk = [
    {"nama": "Produk A", "harga": 100000, "rating": 4.5},
    {"nama": "Produk B", "harga": 50000, "rating": 4.8},
    {"nama": "Produk C", "harga": 200000, "rating": 4.2}
]

sorted_produk = quick_sort(produk, "harga")  # Bisa diganti "rating"
print(sorted_produk)
```
