**Rombel GENAP : Simulasi Tumpukan Barang dalam Gudang**

```python
class Gudang:
    def __init__(self, kapasitas):
        self.kapasitas = kapasitas
        self.tumpukan = []

    def push(self, barang):
        if len(self.tumpukan) < self.kapasitas:
            self.tumpukan.append(barang)
            print(f"Barang '{barang}' ditambahkan ke gudang.")
        else:
            print("Gudang sudah penuh! Tidak bisa menambahkan barang lagi.")

    def pop(self):
        if self.tumpukan:
            barang = self.tumpukan.pop()
            print(f"Barang '{barang}' diambil dari gudang.")
        else:
            print("Gudang kosong! Tidak ada barang untuk diambil.")

    def tampilkan(self):
        if self.tumpukan:
            print("Isi gudang saat ini:", self.tumpukan)
        else:
            print("Gudang kosong!")

# Masukkan dua digit terakhir NPM sebagai kapasitas gudang
npm_akhir = int(input("Masukkan dua digit terakhir NPM Anda: "))
gudang = Gudang(npm_akhir)

while True:
    print("\nMenu:")
    print("1. Tambah Barang (Push)")
    print("2. Ambil Barang (Pop)")
    print("3. Tampilkan Isi Gudang")
    print("4. Keluar")

    pilihan = input("Pilih menu (1-4): ")

    if pilihan == "1":
        barang = input("Masukkan nama barang: ")
        gudang.push(barang)
    elif pilihan == "2":
        gudang.pop()
    elif pilihan == "3":
        gudang.tampilkan()
    elif pilihan == "4":
        print("Program selesai.")
        break
    else:
        print("Pilihan tidak valid, silakan coba lagi.")
```
